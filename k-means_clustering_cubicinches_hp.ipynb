#set initial centroids

```python
a = {}
div = round(len(df)/k)
m={}
print(div)
print(len(df))
```

```python
for i in range(1,k+1):
    a[i] = []
    m[i] = df.loc[div*i-2:div*i-2].values
print(a)
print(m)
```

# functions to find distance and centroid

```python
def dist(med, point):
    #print(med)
    #print(point)
    dist = ma.sqrt((med[0][0]-point[0][0])**2+(med[0][1]-point[0][1])**2)
    return dist
```

```python
def centroid(p, key):
    x = 0
    y=0
    for i in p:
        #print (i)
        x += i[0][0]/len(p)
        y += i[0][1]/len(p)
    #print(x)
    #print(y)
    m[key] = np.array([[x,y]])
```

# training

```python
for _ in range(0,200):
    for key in a:
        a[key] = []
    for i in range(0,len(df)):
        low = None
        for ke in m:
            if low is None:
                low= dist(m[ke], df.loc[i:i].values)
                temp = ke
            elif (dist(m[ke], df.loc[i:i].values) < low):
                low = dist(m[ke], df.loc[i:i].values)
                temp = ke
        a[temp].append(df.loc[i:i].values)
    
    for key in a:
        centroid(a[key], key)
```
